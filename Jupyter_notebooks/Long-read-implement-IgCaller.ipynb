```python
bamfile = pysam.AlignmentFile("E4_mini_IGK.bam","rb")
no_split = pysam.AlignmentFile("No-split_IGK.bam", "wb", template=bamfile)
split = pysam.AlignmentFile("Split_IGK.bam", "wb", template=bamfile)
n = 0
for line in bamfile:
    n +=1
    if line.has_tag("SA"):
        split.write(line)
    else:
        no_split.write(line)

no_split.close()
split.close()
bamfile.close()
```

```python
# Add a tag VR tag to mark variable region
bamfile = pysam.AlignmentFile("Split_IGK.bam","rb")

out_bam = pysam.AlignmentFile("VR_IGK.bam", "wb", template=bamfile)


for line in bamfile:


    #print(line.reference_end)
    bedfile = open("../IgCaller_reference_files/hg38/ucsc/GencodeV29_hg38_IGK_genes_VJ.bed", "r")
    for regions in bedfile:
        column = regions.rstrip("\n").split("\t")



        if int(column[1])-10 <= int(line.reference_start) and int(column[2])+10 >= int(line.reference_start):
            line.tags += [("VR", column[3])]
            print(column[3])
            break
        else:
            line.tags += [("VR", "NA")]
            pass
    out_bam.write(line)

    bedfile.close()
            

    
bamfile.close()
out_bam.close()
```
